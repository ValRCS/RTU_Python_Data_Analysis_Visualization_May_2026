## 4. Markdown analītiskām piezīmēm

Markdown ir vienkāršs veids, kā formatēt tekstu notebook Markdown šūnās.

Labas Markdown piezīmes palīdz nākamajam lasītājam saprast:

- kādu jautājumu mēs uzdodam
- kāpēc palaižam konkrētu koda šūnu
- ko nozīmē izvade
- kādu lēmumu pieņēmām pēc izvades apskatīšanas

Datu analīzē tas ir īpaši svarīgi, jo gala tabula vai diagramma reti ir pietiekama pati par sevi.


### Virsraksti

Izmantojiet virsrakstus, lai sakārtotu notebook.

```markdown
# Galvenais virsraksts
## Galvenā sadaļa
### Mazāka sadaļa
```

Nepārspīlējiet ar virsrakstiem. Notebook vajadzētu būt sakārtotai, nevis sadrumstalotai.


### Rindkopas, treknraksts un iekļautais kods

Markdown piemēri:

```markdown
Šī ir parasta rindkopa.

Šis vārds ir **svarīgs**.

Kolonna `veikals` satur veikalu nosaukumus.
```

Tātad **svarīgs** vārds.
Slīpsvītras *vārds*.
Gan svarīgs gan slīps ***abi divi***.

Iekļautā koda formatējumu lietojiet kolonnu nosaukumiem, failu nosaukumiem un īsiem koda fragmentiem.


### Saraksti

Markdown piemēri:

```markdown
- pirmais elements
- otrais elements
- trešais elements

1. Ielādēt datus.
2. Apskatīt rindas.
3. Uzrakstīt vienu novērojumu.
```

Saraksti ir noderīgi soļiem, pieņēmumiem un novērojumiem.


## Sarakstu piemēri

- biezpiens
* kefīrs (šeit likām zvaigznīti kas viena pati strāda kā slīpsvitra)
- šokolāde

### Numurēts saraksts

1. Izkāpt no gultas
2. Uzvilkt bikses
3. Nomazgāt zobus

### Vienkāršas Markdown tabulas

Markdown piemērs:

```markdown
| Jautājums | Noderīga kolonna |
| --- | --- |
| Kurš veikals parādās visbiežāk? | `veikals` |
| Kurš pirkums maksā visvairāk? | `cena` |
```

Tabulas ir noderīgas īsiem kopsavilkumiem. Lielākas datu tabulas parasti labāk rādīt ar Pandas izvadi.


### Vadīts praktiskais uzdevums: pievienojiet Markdown šūnu

Izveidojiet jaunu Markdown šūnu zem šīs šūnas un uzrakstiet:

```markdown
## Mans pirmais datu jautājums

Es gribu saprast, kurā veikalā ir lielākie kopējie tēriņi.

Noderīgās kolonnas:

- `veikals`
- `cena`
- `daudzums`
```

Pēc tam palaidiet Markdown šūnu un pārbaudiet formatējumu.


## Mans pirmais datu jautājums

Noderīgās kolonnas:

### Biežas Markdown kļūdas

- aizmirsta atstarpe pēc `#` virsrakstā
- pārāk daudz virsrakstu līmeņu
- garas rindkopas tad, kad īsa piezīme būtu skaidrāka
- aizmirsti backticks ap kolonnu nosaukumiem
- aprakstīts kods, bet nav interpretēta izvade

### Jaunas kolonnas izveide

Pirkuma kopējā summa ir:

```text
cena * daudzums
```

Šajā datu kopā tas nozīmē:

```text
cena * daudzums
```


````python
# izveidoju jaunu kolonnua
# pirkumi["summa"] = # te rakstam kas būs rezultāts
pirkumi["summa"] = pirkumi["cena"] * pirkumi["daudzums"]

pirkumi.head()
````

````python
# mēs varam uzstādit arī fiksētu vērtību kādai jaunai (vai vecai kolonna)
# teiksim gads
pirkumi["gads"] = 2026 # visās rindās šai kolonna būs 2026
pirkumi.head()
````

````python
# kā būtu ar kaut kādu rindu dilstošā secība no 100 uz leju pa 1
# so we want new column let's call it priority
# it should start at 100 and go down by 1 per row
pirkumi["prioriteta"] = 100 - pirkumi.index

pirkumi.head()
# atkal redzma ka Pandas darbības notiek kolonās visefektīgāk uzreiz pa visu kolonnu
````

````python
# let's change the prioriteta to be ascending starting from 1000
pirkumi["prioriteta"] = 1000 + pirkumi.index

pirkumi.head()
````

````python
# if I do not want a column I can use drop
pirkumi = pirkumi.drop(columns=["prioriteta"]) # es nometu kolonu nost un rezultāta tabulu saglabāju vecās tabulas vietā, es varēju pirkumi = vietā rakstīt jauna_tabula = tas dotu iespēju veco tabulu saglabāt atmiņā
pirkumi.head()
````

### Kopējo tēriņu kopsavilkums pa veikaliem

Šis ir pirmais nelielais analītiskais rezultāts.

Kods grupē rindas pēc `veikals` un saskaita `summa` vērtības katram veikalam.


````python
pirkumi # full table
````

````python
spending_by_shop = (
    pirkumi
    .groupby("veikals", as_index=False)["summa"]
    .sum() # tā sauktā agregācijas funkcija
    .sort_values("summa", ascending=False) # sakārtojam rezultātus dilstošā secībā
)

spending_by_shop
````

### Nosacījumu zarošanās ar `if`

`if` ļauj Python izvēlēties, ko darīt, balstoties uz nosacījumu.


````python
total = 2.50

if total >= 5: #python kols bieži norāda uz jaunu tvērumu / atkāpi nākoša rindā
    # izpildās tikai šis zars ja lielāks par 5 vai vienāds ar 5
    print("Larger purchase")
else: # pēc izslēgšanas metodes izpildīsies šis zars
    print("Smaller purchase")
# šeit atkal ceļi saiet kopā
print("Vienmēr izpildās neatkarīgi no total vērtības")
````

### Vairāki zari ar `elif`

Izmantojiet `elif`, ja ir vairāk nekā divi iespējamie gadījumi.


````python
total = 3.20

# trīs savstarpēji izslēdzoši ceļi
if total >= 10: # vispirms pārbauda šo
    print("Large")
elif total >= 5:# ta šo - secība svarīga!
    print("Medium")
else: # ja abi pirmās pārbaudes bija False tad šis ceļs
    print("Small")
# atkal visi kopā
````

### Cikli

Cikls (loop) atkārto darbību.

Biežākais iesācēju cikls ir `for` cikls.


````python
print("All shops", shops) # viss saraksts
````

````python
for shop in shops: # atkal atkāpe
    # katrā cikla reizē(iterācijā) shop būs viena vērtība no saraksta
    print(shop)
print("Atceramies ka Pandas parasti ciklus cenšamies neizmantot, jo mums ir vektoru operācijas")
````

Cikli ir noderīgi, bet Pandas bieži ļauj strādāt ar veselām kolonnām uzreiz. Datu analīzē parasti dodam priekšroku DataFrame darbībām, ja tās ir pieejamas.


### Funkcijas

Funkcija ir atkārtoti lietojams koda fragments.

Funkciju izmanto, ja vienu un to pašu loģiku vajag lietot vairāk nekā vienu reizi.


````python
# tātad kaut kas nāk iekšā un kaut kas ārā
def calculate_total(price, quantity):
    return price * quantity
# šeit mums jau ir pieejama funkcija
calculate_total(1.25, 2)
````

````python
# es varu tagad izsaukt šo funkciju vēlreiz un pat saglabāt rezultātu
total = calculate_total(1.25, 2)
print("Mans total", total) # total joprojām ir atmiņā
````

### Funkcija, kas saistīta ar pirkumu datiem

Šī funkcija piešķir pirkumiem lieluma etiķeti.


````python
def purchase_size(total):
    # tātad izdomāju savu loģiku kas būt liels, mazs un vidējs
    if total >= 10:
        return "large"
    elif total >= 5:
        return "medium"
    else:
        return "small"

# te var redzēt kā var izmantot funkciju lai izveidotu jaunu kolonnu
# apply ir pēc būtibas for cikls kas katrai summa šūniņāi pielietos šo funkciju
# un tad saglabās jaunāj kolonnas purchase_size šūniņā
pirkumi["purchase_size"] = pirkumi["summa"].apply(purchase_size) # ievērojiet es padodu purchase_size kā nosaukumu bez purchase_size() kas būtu funkcijas izsaukums
# tātad apply mums kalpoja kā cikls
pirkumi[["pirkums", "veikals", "summa", "purchase_size"]].head()
````

````python
# filtering rows by some rule
# parādīsim tika summas virs 5
pirkumi[pirkumi["summa"] > 5]
# if I wanted to save this view of filtered table
expensive_purchases = pirkumi[pirkumi["summa"] > 5]
expensive_purchases.head()
````
