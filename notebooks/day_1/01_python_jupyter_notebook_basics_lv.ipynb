## 4. Markdown analītiskām piezīmēm

Markdown ir vienkāršs veids, kā formatēt tekstu notebook Markdown šūnās.

Labas Markdown piezīmes palīdz nākamajam lasītājam saprast:

- kādu jautājumu mēs uzdodam
- kāpēc palaižam konkrētu koda šūnu
- ko nozīmē izvade
- kādu lēmumu pieņēmām pēc izvades apskatīšanas

Datu analīzē tas ir īpaši svarīgi, jo gala tabula vai diagramma reti ir pietiekama pati par sevi.


### Virsraksti

Izmantojiet virsrakstus, lai sakārtotu notebook.

```markdown
# Galvenais virsraksts
## Galvenā sadaļa
### Mazāka sadaļa
```

Nepārspīlējiet ar virsrakstiem. Notebook vajadzētu būt sakārtotai, nevis sadrumstalotai.


### Rindkopas, treknraksts un iekļautais kods

Markdown piemēri:

```markdown
Šī ir parasta rindkopa.

Šis vārds ir **svarīgs**.

Kolonna `veikals` satur veikalu nosaukumus.
```

Iekļautā koda formatējumu lietojiet kolonnu nosaukumiem, failu nosaukumiem un īsiem koda fragmentiem.


### Saraksti

Markdown piemēri:

```markdown
- pirmais elements
- otrais elements
- trešais elements

1. Ielādēt datus.
2. Apskatīt rindas.
3. Uzrakstīt vienu novērojumu.
```

Saraksti ir noderīgi soļiem, pieņēmumiem un novērojumiem.


### Vienkāršas Markdown tabulas

Markdown piemērs:

```markdown
| Jautājums | Noderīga kolonna |
| --- | --- |
| Kurš veikals parādās visbiežāk? | `veikals` |
| Kurš pirkums maksā visvairāk? | `cena` |
```

Tabulas ir noderīgas īsiem kopsavilkumiem. Lielākas datu tabulas parasti labāk rādīt ar Pandas izvadi.


### Vadīts praktiskais uzdevums: pievienojiet Markdown šūnu

Izveidojiet jaunu Markdown šūnu zem šīs šūnas un uzrakstiet:

```markdown
## Mans pirmais datu jautājums

Es gribu saprast, kurā veikalā ir lielākie kopējie tēriņi.

Noderīgās kolonnas:

- `veikals`
- `cena`
- `daudzums`
```

Pēc tam palaidiet Markdown šūnu un pārbaudiet formatējumu.


### Biežas Markdown kļūdas

- aizmirsta atstarpe pēc `#` virsrakstā
- pārāk daudz virsrakstu līmeņu
- garas rindkopas tad, kad īsa piezīme būtu skaidrāka
- aizmirsti backticks ap kolonnu nosaukumiem
- aprakstīts kods, bet nav interpretēta izvade


## 5. Pirmais DataFrame darba process ar `pirkumi.csv`

Tagad pārejam no notebook mehānikas uz nelielu datu darba procesu.

Izmantosim `data/pirkumi.csv`, ļoti mazu pirkumu datu kopu ar 12 rindām un 5 kolonnām.

Excel salīdzinājumam:

- CSV fails ir vienkārša teksta tabula
- Pandas DataFrame ir līdzīgs darblapai vai tabulai
- katrai kolonnai ir nosaukums
- katra rinda ir viens ieraksts


### Sagatavošana

Importējam vajadzīgās bibliotēkas.

`pandas` ir galvenā bibliotēka darbam ar tabulām Python vidē. `matplotlib.pyplot` vēlāk tiks izmantots izvēles stabiņu diagrammai (bar chart).


````python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
````

### Pirkumu datu kopas ielāde

Šī šūna ir veidota tā, lai tā darbotos Google Colab vidē un lokāli.

Tā vispirms pārbauda vairākus lokālos ceļus. Ja CSV fails lokāli netiek atrasts, tā ielādē failu no GitHub.


````python
PIRKUMI_URL = "https://raw.githubusercontent.com/ValRCS/RTU_Python_Data_Analysis_Visualization_May_2026/main/data/pirkumi.csv"

possible_paths = [
    Path("data/pirkumi.csv"),
    Path("../data/pirkumi.csv"),
    Path("../../data/pirkumi.csv"),
]

for path in possible_paths:
    if path.exists():
        pirkumi_path = path
        break
else:
    pirkumi_path = PIRKUMI_URL

pirkumi = pd.read_csv(pirkumi_path)
pirkumi.head()
````

### Rindu un kolonnu apskate

`shape` parāda rindu un kolonnu skaitu.

Rezultāts tiek parādīts šādā formā:

```text
(rindas, kolonnas)
```


````python
pirkumi.shape
````

### Kolonnu nosaukumu apskate

Kolonnu nosaukumi ir svarīgi, jo tos izmantojam kodā.


````python
pirkumi.columns
````

### Datu tipu pārbaude

`info()` dod kompaktu pārskatu:

- kolonnu nosaukumi
- neaizpildīto vērtību skaits
- datu tipi

Tas ir līdzīgi kā pārbaudīt, vai Excel kolonnu uztver kā tekstu, skaitli vai datumu.


````python
pirkumi.info()
````

### Skaitlisko kolonnu kopsavilkums

`describe()` dod pamata kopsavilkumus skaitliskajām kolonnām.

Šajā datu kopā skaitliskās kolonnas ir `cena` un `daudzums`.


````python
pirkumi.describe()
````

### Kategoriju skaitīšana

`value_counts()` saskaita, cik bieži katra vērtība parādās kolonnā.

Šeit skaitām pirkumus pa veikaliem.


````python
pirkumi["veikals"].value_counts()
````

### Jaunas kolonnas izveide

Pirkuma kopējā summa ir:

```text
cena * daudzums
```

Šajā datu kopā tas nozīmē:

```text
cena * daudzums
```


````python
pirkumi["summa"] = pirkumi["cena"] * pirkumi["daudzums"]

pirkumi.head()
````

### Kopējo tēriņu kopsavilkums pa veikaliem

Šis ir pirmais nelielais analītiskais rezultāts.

Kods grupē rindas pēc `veikals` un saskaita `summa` vērtības katram veikalam.


````python
spending_by_shop = (
    pirkumi
    .groupby("veikals", as_index=False)["summa"]
    .sum()
    .sort_values("summa", ascending=False)
)

spending_by_shop
````

### Pārveidotā DataFrame saglabāšana

Šī darbība ieraksta jaunu CSV failu ar pievienoto `summa` kolonnu.

Colab vidē fails tiek saglabāts pagaidu izpildes vidē (runtime). VS Code vidē tas tiek saglabāts pašreizējā darba mapē.


````python
pirkumi.to_csv("pirkumi_ar_summu.csv", index=False)

"Saved pirkumi_ar_summu.csv" 
````

### Saglabātā faila pārbaude

Saglabātā faila atkārtota ielasīšana ir vienkāršs veids, kā pārliecināties, ka tas izveidots pareizi.


````python
saved_pirkumi = pd.read_csv("pirkumi_ar_summu.csv")
saved_pirkumi.head()
````

### Salīdzinājumi

Salīdzinājumi atgriež `True` vai `False`.

Tie bieži parādās filtrēšanā un nosacījumu loģikā.


````python
price = 2.50

price > 2
````

````python
shop = "Rimi"

shop == "Rimi" 
````

Biežākie salīdzināšanas operatori:

| Operators | Nozīme |
| --- | --- |
| `==` | vienāds ar |
| `!=` | nav vienāds ar |
| `>` | lielāks nekā |
| `<` | mazāks nekā |
| `>=` | lielāks vai vienāds ar |
| `<=` | mazāks vai vienāds ar |

Bieža kļūda: vērtību salīdzināšanai izmantojiet `==`. Viena vienādības zīme `=` piešķir vērtību mainīgajam.


### Nosacījumu zarošanās ar `if`

`if` ļauj Python izvēlēties, ko darīt, balstoties uz nosacījumu.


````python
total = 7.50

if total >= 5:
    print("Larger purchase")
else:
    print("Smaller purchase")
````

### Vairāki zari ar `elif`

Izmantojiet `elif`, ja ir vairāk nekā divi iespējamie gadījumi.


````python
total = 3.20

if total >= 10:
    print("Large")
elif total >= 5:
    print("Medium")
else:
    print("Small")
````

### Cikli

Cikls (loop) atkārto darbību.

Biežākais iesācēju cikls ir `for` cikls.


````python
for shop in shops:
    print(shop)
````

Cikli ir noderīgi, bet Pandas bieži ļauj strādāt ar veselām kolonnām uzreiz. Datu analīzē parasti dodam priekšroku DataFrame darbībām, ja tās ir pieejamas.


### Funkcijas

Funkcija ir atkārtoti lietojams koda fragments.

Funkciju izmanto, ja vienu un to pašu loģiku vajag lietot vairāk nekā vienu reizi.


````python
def calculate_total(price, quantity):
    return price * quantity

calculate_total(1.25, 2)
````

### Funkcija, kas saistīta ar pirkumu datiem

Šī funkcija piešķir pirkumiem lieluma etiķeti.


````python
def purchase_size(total):
    if total >= 10:
        return "large"
    elif total >= 5:
        return "medium"
    else:
        return "small"

pirkumi["purchase_size"] = pirkumi["summa"].apply(purchase_size)
pirkumi[["pirkums", "veikals", "summa", "purchase_size"]].head()
````
