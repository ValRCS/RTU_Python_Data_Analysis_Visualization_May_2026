## Subchapter 3: Selecting, Filtering, Sorting, and Counting

### Goal

Learn how to focus on selected columns and selected rows.

This is similar to filtering and sorting an Excel table, but the steps are written down so they can be repeated.

### Instructor Demonstration

Select one column by writing its name in square brackets.

```python
titanic["Age"].head()
```

Select several columns by using a list of column names.

```python
titanic[["Name", "Age", "Sex"]].head()
```

Filter rows by writing a condition. The next example keeps passengers whose age is at least 18.

```python
adult_passengers = titanic[titanic["Age"] >= 18]

adult_passengers[["Name", "Age", "Sex", "Pclass"]].head()
```

We can filter using other numeric columns too. This example keeps rows where the fare was greater than 50.

```python
higher_fare_passengers = titanic[titanic["Fare"] > 50]

higher_fare_passengers[["Name", "Fare", "Pclass"]].head()
```

For text or category columns, use `==` to check for an exact match.

```python
first_class_passengers = titanic[titanic["Pclass"] == 1]

first_class_passengers[["Name", "Pclass", "Fare"]].head()
```

Combine conditions with care. Each condition goes inside parentheses, and `&` means "and".

```python
female_first_class = titanic[
    (titanic["Sex"] == "female") &
    (titanic["Pclass"] == 1)
]

female_first_class[["Name", "Sex", "Pclass", "Survived"]].head()
```

Sorting is similar to sorting a table column in Excel.

```python
titanic.sort_values("Fare", ascending=False)[["Name", "Pclass", "Fare"]].head(10)
```

`value_counts()` counts how often each category appears.

```python
titanic["Pclass"].value_counts()
```

```python
titanic["Embarked"].value_counts(dropna=False)
```

## Subchapter 4: Cleaning Missing and Inconvenient Data

### Goal

Learn basic cleaning decisions:

- identify missing values
- decide whether to fill, ignore, or exclude them
- document the choice

Cleaning is not just a technical step. It is part of the analysis.

### Instructor Demonstration

Start by checking missing values again.

```python
missing_report = pd.DataFrame({
    "missing_rows": titanic.isna().sum(),
    "missing_percent": (titanic.isna().mean() * 100).round(1)
})

missing_report.sort_values("missing_rows", ascending=False)
```

For this beginner analysis, we will make three simple decisions:

1. Keep `Cabin` out of the main analysis because most values are missing.
2. Fill missing `Age` values with the median age in a new column called `Age_filled`.
3. Drop the few rows with missing `Embarked` values for summaries that use that column.

We create a cleaned working copy instead of changing the original table.

```python
titanic_clean = titanic.dropna(subset=["Embarked"]).copy()

median_age = titanic_clean["Age"].median()
median_age
```

```python
titanic_clean["Age_filled"] = titanic_clean["Age"].fillna(median_age)

titanic_clean[["Age", "Age_filled", "Embarked"]].head(10)
```

Compare missing values before and after the cleaning step.

```python
before_after_missing = pd.DataFrame({
    "original_missing": titanic.isna().sum(),
    "cleaned_missing": titanic_clean.isna().sum()
})

before_after_missing.sort_values("original_missing", ascending=False)
```

Notice that `Age` still has missing values, but `Age_filled` does not. This keeps both pieces of information:

- the original `Age` column
- a practical filled version for simple summaries

```python
titanic_clean[["Age", "Age_filled"]].describe()
```

### Guided Practice

```python
titanic_clean["Embarked"].value_counts(dropna=False)
```

### Check Your Understanding

Answer in plain language:

- Why did we create `Age_filled` instead of replacing `Age` directly?
- Why is `Cabin` not a good main analysis column here?
- What cleaning decision did we make for `Embarked`?

### Common Mistake

Missing values are not the same as zero.

If age is missing, that does not mean the passenger was 0 years old. It means the age was not available in the dataset.

## Subchapter 5: Grouping, Summarizing, and Mini Report Prep

### Goal

Learn how to move from individual rows to summary tables.

In Excel, this is similar to creating a Pivot Table. In Pandas, the main tools are `groupby()` and `pivot_table()`.

### Instructor Demonstration

First, calculate the overall survival rate. `Survived` is coded as:

- `1` for survived
- `0` for did not survive

The mean of a 0/1 column is the share of rows with value 1.

```python
overall_survival_rate = titanic_clean["Survived"].mean()

round(overall_survival_rate, 3)
```

Now compare survival rates by passenger sex.

```python
survival_by_sex = titanic_clean.groupby("Sex")["Survived"].mean().round(3).reset_index()

survival_by_sex
```

Compare survival rates by passenger class.

```python
survival_by_class = titanic_clean.groupby("Pclass")["Survived"].mean().round(3).reset_index()

survival_by_class
```

Compare average fare by passenger class.

```python
fare_by_class = titanic_clean.groupby("Pclass")["Fare"].mean().round(2).reset_index()

fare_by_class
```

We can summarize two grouping columns at the same time. This is close to a Pivot Table layout.

```python
survival_pivot = pd.pivot_table(
    titanic_clean,
    values="Survived",
    index="Sex",
    columns="Pclass",
    aggfunc="mean"
).round(3)

survival_pivot
```

A simple chart can make a summary easier to compare. Day 2 focuses more deeply on charts, but we can introduce one readable chart here.

```python
plt.figure(figsize=(8, 5))

sns.barplot(data=survival_by_class, x="Pclass", y="Survived")

plt.title("Survival Rate Was Higher in First Class")
plt.xlabel("Passenger class")
plt.ylabel("Survival rate")
plt.ylim(0, 1)
plt.show()
```
