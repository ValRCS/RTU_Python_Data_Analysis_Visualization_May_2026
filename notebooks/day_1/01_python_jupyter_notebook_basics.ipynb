## 1. Why Computational Notebooks Exist

A computational notebook combines four things in one document:

- explanatory text
- runnable code
- visible output
- interpretation of results

That makes notebooks useful for reproducible analysis. You can show not only the final answer, but also how the answer was produced.

This is different from a plain script, where the code is usually separate from the explanation. It is also different from a spreadsheet, where many steps may be hidden inside formulas, filters, or manual actions.


### Notebooks Compared with Other Tools

| Tool | Good for | Main limitation |
| --- | --- | --- |
| Spreadsheet | familiar tables, formulas, quick filtering, charts | repeated manual steps can be hard to document |
| Python script | automated repeatable code | less friendly for teaching, notes, and step-by-step interpretation |
| Computational notebook | code, output, notes, and interpretation together | cells can be run out of order if we are not careful |

In this course, notebooks are used as analytical workbooks. They are not just places to write code. They are places to document a complete analysis workflow.

## 2. Jupyter Notebook Basics

Jupyter notebooks are made from cells.

The two cell types we use most are:

- Markdown cells: formatted text for explanations, questions, and conclusions
- Code cells: Python instructions that can be run

A notebook also has a kernel. The kernel is the Python process that remembers variables while the notebook is running.


### Code Cells Remember Variables

Run the next cell. It creates a variable called `name`.


````python
name = "Anna"
name
````

The next cell uses the variable created above. This works because the notebook kernel remembers it.


````python
greeting = "Hello, " + name
greeting
````

### Cell Order Matters

If you restart the runtime and run only the second cell, Python will not know what `name` means.

This is one of the most common notebook mistakes: the visible order of cells and the actual run order can become different.

Practical habit:

- use **Runtime > Run all** before trusting final results
- keep setup and data-loading cells near the top
- avoid depending on hidden manual steps


### Code Cell Output

A code cell displays the last expression automatically. You can also use `print()` when you want to show several things.


````python
print("Price:", price)
print("Quantity:", quantity)
print("Total:", total)
````

## 4. Markdown for Analytical Notes

Good Markdown notes help future readers understand:

## 5. First DataFrame Workflow with `pirkumi.csv`

Now we move from notebook mechanics to a small data workflow.

We will use `data/pirkumi.csv`, a tiny purchase dataset with 12 rows and 5 columns.

In Excel terms:

- a CSV file is a plain-text table
- a Pandas DataFrame is similar to a worksheet or table
- each column has a name
- each row is one record


### Setup

We import the libraries we need.

`pandas` is the main library for working with tables in Python. `matplotlib.pyplot` is used later for the optional bar chart.


````python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
````

### Load the Purchases Dataset

This cell is designed to work in Google Colab and locally.

It checks several local paths first. If the CSV file is not found locally, it loads the file from GitHub.


````python
PIRKUMI_URL = "https://raw.githubusercontent.com/ValRCS/RTU_Python_Data_Analysis_Visualization_May_2026/main/data/pirkumi.csv"

possible_paths = [
    Path("data/pirkumi.csv"),
    Path("../data/pirkumi.csv"),
    Path("../../data/pirkumi.csv"),
]

for path in possible_paths:
    if path.exists():
        pirkumi_path = path
        break
else:
    pirkumi_path = PIRKUMI_URL

pirkumi = pd.read_csv(pirkumi_path)
pirkumi.head()
````

### Inspect Rows and Columns

`shape` tells us the number of rows and columns.

The result is shown as:

```text
(rows, columns)
```


````python
pirkumi.shape
````

### View Column Names

Column names are important because we use them in code.


````python
pirkumi.columns
````

### Inspect Data Types

`info()` gives a compact overview:

- column names
- number of non-empty values
- data types

This is similar to checking whether Excel treats a column as text, number, or date.


````python
pirkumi.info()
````

### Summarize Numeric Columns

`describe()` gives basic summaries for numeric columns.

For this dataset, the numeric columns are `cena` and `daudzums`.


````python
pirkumi.describe()
````

### Count Categories

`value_counts()` counts how often each value appears in a column.

Here we count purchases by shop.


````python
pirkumi["veikals"].value_counts()
````

### Create a New Column

The total amount for a purchase is:

```text
price * quantity
```

In this dataset, that means:

```text
cena * daudzums
```


````python
pirkumi["summa"] = pirkumi["cena"] * pirkumi["daudzums"]

pirkumi.head()
````

### Summarize Total Spending by Shop

This is a first small analytical result.

The code groups rows by `veikals` and adds the `summa` values for each shop.


````python
spending_by_shop = (
    pirkumi
    .groupby("veikals", as_index=False)["summa"]
    .sum()
    .sort_values("summa", ascending=False)
)

spending_by_shop
````

### Save the Modified DataFrame

This writes a new CSV file with the added `summa` column.

In Colab, the file is saved in the temporary runtime. In VS Code, it is saved in the current working folder.


````python
pirkumi.to_csv("pirkumi_ar_summu.csv", index=False)

"Saved pirkumi_ar_summu.csv" 
````

### Check the Saved File

Reading the saved file back is a simple way to confirm that it was created correctly.


````python
saved_pirkumi = pd.read_csv("pirkumi_ar_summu.csv")
saved_pirkumi.head()
````

## 6. Python Basics for Reading Notebook Code

The goal of this section is not to become a software developer.

The goal is to understand enough Python to read, run, and lightly modify data-analysis notebook cells.


### Values and Variables

A variable is a name that stores a value.

In notebooks, variables let us reuse previous results.


````python
price = 1.25
quantity = 2
product = "milk"

price, quantity, product
````

### Basic Data Types

Common Python data types:

| Type | Meaning | Example |
| --- | --- | --- |
| `int` | whole number | `3` |
| `float` | decimal number | `2.50` |
| `str` | text | `"Rimi"` |
| `bool` | true or false | `True` |


````python
type(price), type(quantity), type(product), type(price > 2)
````

### Text Values

Text values are called strings.

Strings must be written inside quotation marks.


````python
shop = "Rimi"
message = "Purchase was made at " + shop

message
````

### Lists

A list stores several values in order.

Lists are useful when we want to keep several related items together.


````python
shops = ["Rimi", "Maxima", "Lidl"]

shops
````

We can select one item from a list by position.

Python starts counting from 0.


````python
shops[0]
````

### Dictionaries

A dictionary stores pairs of keys and values.

This is useful when each value has a label.


````python
purchase = {
    "pirkums": "piens",
    "cena": 1.25,
    "daudzums": 2,
    "veikals": "Rimi",
}

purchase
````

````python
purchase["cena"]
````

### Comparisons

Comparisons return `True` or `False`.

These are common in filtering and conditional logic.


````python
price = 2.50

price > 2
````

````python
shop = "Rimi"

shop == "Rimi" 
````

Common comparison operators:

| Operator | Meaning |
| --- | --- |
| `==` | equal to |
| `!=` | not equal to |
| `>` | greater than |
| `<` | less than |
| `>=` | greater than or equal to |
| `<=` | less than or equal to |

Common mistake: use `==` when comparing values. A single `=` assigns a value to a variable.


### Branching with `if`

`if` lets Python choose what to do based on a condition.


````python
total = 7.50

if total >= 5:
    print("Larger purchase")
else:
    print("Smaller purchase")
````

### Multiple Branches with `elif`

Use `elif` when there are more than two possible cases.


````python
total = 3.20

if total >= 10:
    print("Large")
elif total >= 5:
    print("Medium")
else:
    print("Small")
````

### Loops

A loop repeats an action.

The most common beginner loop is a `for` loop.


````python
for shop in shops:
    print(shop)
````

Loops are useful, but Pandas often lets us work with whole columns at once. In data analysis, we usually prefer DataFrame operations when they are available.


### Functions

A function is a reusable piece of code.

Use a function when the same logic should be applied more than once.


````python
def calculate_total(price, quantity):
    return price * quantity

calculate_total(1.25, 2)
````

### A Function Connected to the Purchases Data

This function labels purchases by size.


````python
def purchase_size(total):
    if total >= 10:
        return "large"
    elif total >= 5:
        return "medium"
    else:
        return "small"

pirkumi["purchase_size"] = pirkumi["summa"].apply(purchase_size)
pirkumi[["pirkums", "veikals", "summa", "purchase_size"]].head()
````
